```python
import pandas as pd
import os
print(os.getcwd())

df = df = pd.read_csv("C:/Users/victo/Downloads/train.csv")

df.head(10)
df.columns
df.describe()
df.isnull().sum()

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.histplot(df["Sales"], bins=100, kde=True)  # Falls es "Sales" nicht gibt, ersetze durch eine andere Spalte
plt.title("Verteilung der Verkaufswerte")
plt.show()
df["Category"].value_counts()
df.head(10)
df.sort_values(by="Sales", ascending=False).head(10)  # Zeigt die 10 teuersten Produkte

sns.barplot(x=df.groupby("Category")["Sales"].sum().sort_values(ascending=False).index, 
            y=df.groupby("Category")["Sales"].sum().sort_values(ascending=False).values)
plt.xticks(rotation=45)
plt.title("Umsatz pro Kategorie")
plt.xlabel("Kategorie")
plt.ylabel("Gesamtumsatz ()")
plt.show()

top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

sns.barplot(x=top_products.values, y=top_products.index)
plt.title("Top 10 Produkte nach Umsatz")
plt.xlabel("Gesamtumsatz ()")
plt.ylabel("Produkt")
plt.show()

top_customers = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)

sns.barplot(x=top_customers.values, y=top_customers.index)
plt.title("Top 10 Kunden nach Umsatz")
plt.xlabel("Gesamtumsatz ()")
plt.ylabel("Kunde")
plt.show()

df.head(10)
import os
print(os.getcwd())
```
